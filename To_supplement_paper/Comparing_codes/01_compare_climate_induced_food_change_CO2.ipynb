```python
import pandas as pd
import plotnine
from plotnine import *
```

```python
# read data
prod_with_CO2 = pd.read_csv('./data/climate_withCO2.csv')
prod_without_CO2 = pd.read_csv('./data/climate_withoutCO2.csv')

# add CO2 flag
prod_with_CO2['CO2'] = 'YES'
prod_without_CO2['CO2'] = 'NO'

# concat data
prod_CO2 = pd.concat([prod_with_CO2,prod_without_CO2],0)
```

```python
# plot
# plot the yearbook/GAEZ/GYGA data
plotnine.options.figure_size = (7, 5.3)
plotnine.options.dpi = 100

g = (ggplot(prod_CO2) +
 geom_ribbon(aes(x='final_year',
          ymin='accum_climate_bot',
          ymax='accum_climate_top',fill='CO2'),alpha=0.1) +

 geom_line(aes(x='final_year',y='accum_climate',color='CO2')) +

 theme_bw(base_size=13) + 
#  coord_cartesian(ylim=(-120,200)) +
#  theme(legend_title=element_blank(),legend_key_size=12,legend_text=element_text(size=10)) +
 ylab('Cumulative Contribution (million t)') +
 xlab(' ') 
)

# show fig
g
```
